```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Caricamento dei dataset integrati
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
penguins = sns.load_dataset("penguins")
```

2. Scatterplot: relazione tra total_bill e tip, differenziati per giorno (day)


```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("relazione conto totale e mancia per giorno")
plt.show()
```

2. Lineplot: andamento passeggeri nel tempo


```python
passeggeri_anno = flights.groupby("year")["passengers"].sum().reset_index()

sns.lineplot(data= passeggeri_anno, x="year", y="passengers", marker="o")
plt.title("passeggeri per anno")
plt.show()
```

- Fase 4: Grafici di Distribuzione 
    - Create un histogram della variabile total_bill dal dataset tips, aggiungendo anche una curva KDE.
    - Realizzate un pairplot del dataset penguins, utilizzando species come parametro hue.


```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("distribuzione totale del conto con curva")
plt.show()
```

Pairplot del dataset penguins con species come hue

```python
sns.pairplot(data=penguins, hue="species")
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
sns.set_theme(style="darkgrid")

sns.histplot(data=diamonds,
            x="price",
            kde=True, 
            bins=30,
            color="gold")
plt.title("distribuzione prezzo dei diamanti")
plt.xlabel("prezzo")
plt.ylabel("frequenza")
plt.grid(True, linestyle='--')
plt.show()
```
