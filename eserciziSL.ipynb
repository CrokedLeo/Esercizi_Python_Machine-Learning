```python
df = pd.read_csv('fabbisogno calorico.csv', index_col=0)
print(df.head())
```

```python
print(y.min())
```

```python
X = df[['eta']] #input
y = df['calorie'] #output
```

```python
X_train, X_test, y_train ,y_test= train_test_split(X,y)
```

```python
model= LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse:.2f}")
print(f"R¬≤ Score: {r2:.2f}")
print(f"Intercetta: {model.intercept_:.2f}")
print(f"Coefficiente (pendenza): {model.coef_[0]:.2f}")
```

### Grafico retta e dati

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="#10ec6b", label='Dati reali')
plt.plot(X, model.predict(X), color='black', label='Retta di regressione')
plt.title('Et√† vs Fabbisogno calorico giornaliero')
plt.xlabel('Et√† (anni)')
plt.ylabel('Fabbisogno calorico (kcal)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Confronto test e predizione
